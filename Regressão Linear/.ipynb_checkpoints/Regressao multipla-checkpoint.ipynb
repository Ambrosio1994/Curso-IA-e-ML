```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
data = fetch_california_housing()
data
```

```python
df = pd.DataFrame(data.data)
df.columns = data.feature_names
df['MedHouseVal'] = data.target
df
```

```python
corr = df.corr()
corr
```

```python
previsores = df[['MedInc', 'AveRooms']].values
classe = df.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores, classe, test_size=0.3, random_state=42)
```

```python
modelo = LinearRegression()
modelo.fit(x_train, y_train)
```

```python
pred = modelo.predict(x_test)
pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(pred, y_test)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Calculando o Erro QuadrÃ¡tico MÃ©dio (MSE)
mse = mean_squared_error(y_test, pred)
print(f"MSE: {mse}")

# Calculando a Raiz do Erro QuadrÃ¡tico MÃ©dio (RMSE)
rmse = mean_squared_error(y_test, pred, squared=False)
print(f"RMSE: {rmse}")

# Calculando o Erro Absoluto MÃ©dio (MAE)
mae = mean_absolute_error(y_test, pred)
print(f"MAE: {mae}")

# Calculando o Coeficiente de DeterminaÃ§Ã£o (RÂ²)
r_squared = r2_score(y_test, pred)
print(f"RÂ²: {r_squared}")
```
