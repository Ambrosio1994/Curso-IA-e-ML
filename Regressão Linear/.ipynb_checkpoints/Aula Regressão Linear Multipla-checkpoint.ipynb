```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
```

```python
df = pd.read_csv('mt_cars.csv')
df
```

```python
df.shape
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
df.head()
```

```python
x = df.iloc[:, 2].values # dist
y = df.iloc[:, 0].values # km/l
```

```python
correlação = np.corrcoef(x, y)
correlação
```

```python
x = x.reshape(-1, 1)
```

```python
modelo = LinearRegression()
modelo.fit(x, y)
```

```python
modelo.intercept_
```

```python
modelo.coef_
```

```python
modelo.score(x, y)
```

```python
previsão = modelo.predict(x)
previsão
```

```python
modelo_ajustado = sm.ols(formula = 'mpg ~ disp', data = df)
modelo_treinado = modelo_ajustado.fit()
modelo_treinado.summary()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.plot(x, previsão, color = 'red')
```

```python
modelo.predict([[500]])
```
