```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from yellowbrick.regressor import ResidualsPlot
```

```python
df = pd.read_csv('cars.csv')
df
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df
```

```python
previsor = df.iloc[:, 1].values
classe = df.iloc[:, 0].values
```

```python
previsor
```

```python
classe
```

```python
corr = df.corr()
corr
```

```python
previsor = previsor.reshape(-1, 1)
previsor
```

```python
modelo = LinearRegression()
modelo.fit(previsor, classe)
```

```python
plt.scatter(previsor, classe) # dados
plt.plot(previsor, modelo.predict(previsor), color = 'r') # linha de melhor ajuste
plt.show()
```

```python
modelo.predict([[22]])
```

```python
residuos = ResidualsPlot(modelo)
residuos.fit(previsor, classe)
residuos.poof()
```
