```python
import spacy
import re
import string
```

```python
pln = spacy.load("pt_core_news_sm")

pln
```

```python
documento = pln('inteligente inteligencia inteligentes corrida correr correndo correstes')
for token in documento:
    print(token.text, token.lemma_)
```

```python
def processamento_lematizado(texto):
    texto = texto.lower()
    texto = re.sub(r"\s+", ' ', texto)

    documento = pln(texto)
    tokens = []
    for token in documento:
        tokens.append(token.lemma_)

    stop_words = spacy.lang.pt.STOP_WORDS
    tokens = [palavra for palavra in tokens if palavra not in stop_words and palavra not in string.punctuation]
    
    texto_formatado = " ".join([str(elemento) for elemento in tokens if not elemento.isdigit()])

    return texto_formatado
```

```python
texto_original = """A inteligência artificial é a inteligência similar à humana. 
                    Definem como o estudo de agente artificial com inteligência. 
                    Ciência e engenharia de produzir máquinas com inteligência. 
                    Resolver problemas e possuir inteligência. 
                    Relacionada ao comportamento inteligente. 
                    Construção de máquinas para raciocinar. 
                    Aprender com os erros e acertos. 
                    Inteligência artificial é raciocinar nas situações do cotidiano."""
```

```python
processamento_lematizado(texto_original)
```
