```python
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
corpus = PlaintextCorpusReader('Arquivos/', '.*\.txt',  encoding = 'ISO-8859-1')
corpus
```

```python
arquivos = corpus.fileids()
arquivos[0]
```

```python
arquivos[0:10]
```

```python
texto = corpus.raw('1.txt')
texto
```

```python
palavras = corpus.words()
palavras[170]
```

```python
todo_texto = corpus.raw()
```

```python
stops = stopwords.words('english')
stops
```

```python
mapa_cores = ListedColormap(['orange', 'green', 'red', 'magenta'])
nuvem = WordCloud(background_color = 'white',
                  colormap = mapa_cores,
                  stopwords = stops,
                  max_words = 100)
nuvem.generate(todo_texto)
plt.imshow(nuvem)
```

```python
palavras2 = [p for p in palavras if p not in stops]
```

```python
palavras3 = [p for p in palavras2 if p not in string.punctuation]
```

```python
frequencia = nltk.FreqDist(palavras3)
frequencia
```

```python
top = frequencia.most_common(100)
top
```
