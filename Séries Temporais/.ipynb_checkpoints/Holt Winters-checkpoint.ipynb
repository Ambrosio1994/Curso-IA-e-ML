```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import rcParams
rcParams['figure.figsize']= 18,10
import numpy as np
from datetime import datetime
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
df = pd.read_csv('AirPassengers.csv', parse_dates = ['Month'], index_col = 'Month')
df
```

```python
df.index.freq = 'MS' # Definindo a frequencia com que os dados serão analisados

modelo = ExponentialSmoothing(df, seasonal_periods=12, trend = 'additive',  seasonal = 'additive', use_boxcox = True).fit()
```

```python
modelo.forecast(24).plot(style = '--', marker = 'o', color = 'black', legend = True)
plt.show()
```

```python
modelo.fittedvalues.plot(style = '--', color = 'r')
modelo.forecast(24).plot(style = '-', marker = 'o', color = 'b', legend = True)
plt.show()
```
