```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('avocado.csv', parse_dates = ['Date'])
df.head()
```

```python
summr = df.groupby('Date')['Total Volume'].mean().reset_index()
summr.head()
```

```python
df.index
```

```python
fig, ax = plt.subplots(1,1, figsize = (12, 8))
summr.set_index('Date').plot(ax = ax, marker = 'd', linestyle = '-', color = 'blue')
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize = (12, 8))
summr = summr.set_index('Date')

summr['month'] = summr.index.month

agrupado = summr.groupby('month')['Total Volume'].agg(['mean', 'std', 'median', 'min', 'max'])

fig, ax = plt.subplots()
agrupado.plot(ax=ax, marker='d', linewidth=2)
ax.set_xlabel('Ms')
plt.show()
```
