```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.pyplot import rcParams
rcParams['figure.figsize']= 18,10
import numpy as np
from datetime import datetime
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
df = pd.read_csv('AirPassengers.csv', parse_dates = ['Month'], index_col = 'Month')
df.head()
```

```python
df.index.freq = 'MS' # Definindo a frequencia com que os dados serão analisados

# 80% de peso para os dddos mais recentes,20% para os mais antigos. Tendência ter peso de 20%
modelo = Holt(df, damped_trend = True).fit(smoothing_level = 0.8, smoothing_trend = 0.2) 

modelo1 = modelo.forecast(12) # Previsão de 12 intervalos de dados futuros
```

```python
modelo1.plot( marker = '.', color = 'blue', legend = True)
modelo.fittedvalues.plot(marker = '.', color = 'red')
plt.show()
```
