```python
import pandas as pd
from sklearn.model_selection import train_test_split
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from yellowbrick.classifier import ConfusionMatrix
```

```python
df = pd.read_csv('Credit.csv')
df
```

```python
previsores = df.iloc[:,:-1].values
classe = df.iloc[:,-1].values
```

```python
# Criando uma instância do LabelEncoder
labelencoder = LabelEncoder()

# Colunas para codificar
columns_to_encode = [0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19]

# Aplicando a transformação para as colunas selecionadas
for col in columns_to_encode:
    previsores[:, col] = labelencoder.fit_transform(previsores[:, col])
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores, classe, test_size = 0.3, random_state = 42)
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(x_train, y_train)
```

```python
pred = modelo.predict(x_test)
pred
```

```python
verificação = ConfusionMatrix(DecisionTreeClassifier())
verificação.fit(x_train, y_train)
verificação.score(x_test, y_test)
verificação.show()
```
