```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AirPassengers.csv', parse_dates = ['Month'], index_col = 'Month')
df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 12,5


plt.plot(df)
plt.show()
```

```python
from pmdarima.arima import auto_arima


stepwise_model1 = auto_arima(df, stat_p = 1, start_q = 1, start_d = 0, start_P = 0, max_p = 6, max_q = 6, m= 12,
                            seasonal = True, trace = True, stepwise = True)
```

```python
stepwise_model = auto_arima(df, stat_p = 1, start_q = 1, start_d = 0, start_P = 0, max_p = 6, max_q = 6, m= 12,
                            seasonal = True, trace = True, stepwise = False)
```

```python
print(stepwise_model1.aic())
print(stepwise_model.aic())
```

```python
treino = df.loc['1949-01-01':'1959-12-01']
teste = df.loc['1960-01-01':]
```

```python
treino
```

```python
teste
```

```python
stepwise_model.fit(treino)
```

```python
future_forecast = stepwise_model.predict(12)
future_forecast
```

```python
future_forecast = pd.DataFrame(future_forecast, index = teste.index, columns = ['#Passengers'])
future_forecast
```

```python
pd.concat([teste, future_forecast], axis = 1).plot()
plt.show()
```

```python
pd.concat([df, future_forecast], axis = 1).plot()
plt.show()
```
