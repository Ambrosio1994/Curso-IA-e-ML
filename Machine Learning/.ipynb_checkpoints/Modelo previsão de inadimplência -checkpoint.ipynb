```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,accuracy_score
from yellowbrick.classifier import ConfusionMatrix
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df = pd.read_csv('Anonymize_Loan_Default_data.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.dropna()
```

```python
previsores = df.iloc[:, :-1].values
classe = df.iloc[:, -1].values
```

```python
colunas_nao_numericas = df.select_dtypes(exclude=['number']).columns
```

```python
# Crie uma instância do LabelEncoder
labelencoder = LabelEncoder()

columns_to_encode = colunas_nao_numericas

# Índices das colunas não numéricas a serem codificadas
columns_to_encode_indices = [df.columns.get_loc(col) for col in columns_to_encode]

# Aplicando a transformação para as colunas selecionadas
for col in columns_to_encode_indices:
    previsores[:, col] = labelencoder.fit_transform(previsores[:, col])
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores, classe,
                                                    test_size = 0.3, random_state = 0)
```

```python
from sklearn.impute import SimpleImputer

# Preenchendo valores ausentes usando a média
imputer = SimpleImputer(strategy='mean')
x_train = imputer.fit_transform(x_train)

# Criando e treinando o modelo
model = SVC()
model.fit(x_train, y_train)
```

```python
x_test = imputer.transform(x_test)


# Fazendo previsões com os dados de teste tratados
previsao = model.predict(x_test)
```

```python
previsao
```

```python
verificação = ConfusionMatrix(SVC())
verificação.fit(x_train, y_train)
verificação.score(x_test, y_test)
verificação.show()
```

```python
taxa_acerto = accuracy_score(y_test, previsao)
taxa_acerto
```
