```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 15, 7
```

```python
df = pd.read_csv('AirPassengers.csv', parse_dates = [0], header = 0, index_col = 0).squeeze('columns')
df
```

```python
# Transformando a série estacionaria com 12 períodos.
difer = df.diff(12)
difer.head(13)
```

```python
difer.plot()
plt.show()
```

```python
# Removendo os 12 primeiros registros

difer = difer[12:]
```

```python
difer.head(13)
difer.plot()
plt.show()
```

```python
# Criando um DataFrame tabular com colunas defasadas (lagged) e a coluna alvo
dataframe = pd.DataFrame({'t-'+str(i): difer.shift(i) for i in range(12, 0, -1)})

# Adicionando a coluna alvo 't' com os valores originais da série temporal
dataframe['t'] = difer.values

# Removendo as primeiras 12 linhas que contêm valores NaN devido ao deslocamento
dataframe = dataframe[12:]
```

```python
dataframe
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Convertendo o DataFrame em uma matriz numpy para separar features (x) e rótulos (y)
array = dataframe.values

# Selecionando todas as linhas da primeira coluna do array,
# que contém as features (variáveis independentes)
x = array[:, 0:-1]

# Selecionando todas as linhas da última coluna do array,
# que contém os rótulos (variável dependente)
y = array[:, -1]
```

```python
# Criando um modelo de regressão usando o algoritmo Random Forest
# com 500 árvores de decisão e definindo a semente para a aleatoriedade como 1
model = RandomForestRegressor(n_estimators=500, random_state=1)

# Treinando o modelo com os dados de entrada (features x) e saída correspondente (rótulos y)
model.fit(x, y)
```

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt

# Obtendo os nomes das colunas do DataFrame, exceto a última coluna (rótulo)
nome = dataframe.columns.values[0:-1]

# Criando uma lista de valores para os marcadores no eixo x
ticks = [i for i in range(len(nome))]

# Criando um gráfico de barras com as importâncias das features do modelo
plt.bar(ticks, model.feature_importances_)

# Definindo os rótulos do eixo x com os nomes das features
plt.xticks(ticks, nome)

# Exibindo o gráfico
plt.show()
```
