```python
import torch
from torch import nn, optim
import torchvision.transforms as transforms
import torchvision.datasets as dsets
import matplotlib.pyplot as plt

torch.manual_seed(1)
```

```python
class Softmax(nn.Module):

    def __init__(self, in_size, out_size):

        super(Softmax, self).__init__()

        self.linear=nn.Linear(in_size, out_size)

    def forward(self, x):

        out=self.linear(x)

        return out
```

```python
train_dataset = dsets.MNIST(root="./data", train=True, download=True, transform=transforms.ToTensor())
val_dataset = dsets.MNIST(root="./data", train=False, download=True, transform=transforms.ToTensor())
```

```python
input_dim = 28*28
output_dim = 10

model = Softmax(input_dim, output_dim)
parameters = model.parameters()

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(parameters, lr=0.01)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=100, shuffle=True)
val_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=5000, shuffle=True)
```

```python
EPOCHS = 100
ACC = []

for epoch in range(EPOCHS):
  for image, label in train_loader:
    image = image.view(-1, 28*28)

    # Forward pass
    outputs = model(image)
    loss = criterion(outputs, label)

    # Backward pass e otimização
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

  correct = 0
  total = 0

  for image, label in val_loader:
    image = image.view(-1, 28*28)
    outputs = model(image)
    _, yhat = torch.max(outputs.data, 1)
    correct += (yhat == label).sum().item()
    total += label.size(0)

  acc = correct / total
  ACC.append(acc)


  print(f'Epoch [{epoch+1}/{EPOCHS}], Accuracy: {acc:.4f}')
print('Training complete')
```
