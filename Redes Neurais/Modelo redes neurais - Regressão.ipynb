```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('autos.csv', encoding = 'ISO-8859-1')
df
```

```python
df.columns
```

```python
df = df.drop(['dateCrawled', 'name', 'seller', 'offerType', 'dateCreated', 'nrOfPictures',
              'postalCode', 'lastSeen'] , axis = 1)
df
```

```python
for coluna in df.columns:
    if df[coluna].dtype == 'float64':
        if df[coluna].isna().any():
            media = df[coluna].mean()
            df[coluna].fillna(media, inplace=True)

for coluna in df.columns:
    if df[coluna].dtype == 'object':
        if df[coluna].isna().any():
            moda = df[coluna].mode().iloc[0]
            df[coluna].fillna(moda, inplace=True)
```

```python
df
```

```python
df.loc[df['price'] <= 10] = df['price'].mean()
df = df[df['price'] < 350000]
df
```

```python
previsores = df.iloc[:, 1:13].values
classe = df.iloc[: , 0].values
previsores
```

```python
labelencoder = LabelEncoder()

for coluna in range(previsores.shape[1]):
    if previsores[:, coluna].dtype == object:
        
        previsores[:, coluna] = previsores[:, coluna].astype(str)
        
        previsores[:, coluna] = labelencoder.fit_transform(previsores[:, coluna])
```

```python
from sklearn.compose import ColumnTransformer

column_transformer = ColumnTransformer([
    ('categorical_features', OneHotEncoder(), [0, 1, 3, 5, 8, 9, 10]),
])

previsores = column_transformer.fit_transform(previsores)
```

```python
previsores.shape[1]
```

```python
modelo = Sequential()

modelo.add(Dense(158, activation = 'relu', input_dim = previsores.shape[1]))
modelo.add(Dropout(0.3))

modelo.add(Dense(158, activation = 'relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(158, activation = 'relu'))

modelo.add(Dense(1, activation = 'linear'))
```

```python
modelo.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['mean_absolute_error'])
modelo.fit(previsores, classe, batch_size = 300, epochs = 100)
```
