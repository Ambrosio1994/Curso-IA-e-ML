```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df['target'] = data.target
df
```

```python
previsores = df.iloc[:,:-1].values
classe = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(previsores, classe, test_size = 0.3, random_state = 25)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
modelo = Sequential()

modelo.add(Dense(16, kernel_initializer='random_uniform', input_dim=x_train.shape[1], activation='relu'))
modelo.add(Dropout(0.1))

modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
modelo.add(Dropout(0.1))

modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
modelo.add(Dropout(0.1))

modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
modelo.add(Dropout(0.1))

modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))

modelo.add(Dense(1, activation='sigmoid'))
```

```python
#otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5,
#                                      beta_1=0.9, beta_2=0.999, epsilon=1e-07, amsgrad=False)

#modelo.compile(optimizer = otimizador, loss ='binary_crossentropy', metrics = ['accuracy'])
modelo.compile(optimizer = 'adam', loss ='binary_crossentropy', metrics = ['accuracy'])

modelo.fit(x_train, y_train, epochs = 150, batch_size = 10, validation_data = (x_test, y_test))
```

```python
pred = modelo.predict(x_test)
pred = (pred > 0.5)
pred
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)
cm
```

```python
from sklearn.model_selection import cross_val_score
import tensorflow as tf
from scikeras.wrappers import KerasClassifier
```

```python
def CriarRede():
    modelo = Sequential()

    modelo.add(Dense(16, kernel_initializer='random_uniform', input_dim=x_train.shape[1], activation='relu'))
    modelo.add(Dropout(0.1))
    
    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    modelo.add(Dropout(0.1))
    
    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    modelo.add(Dropout(0.1))
    
    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    modelo.add(Dropout(0.1))
    
    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    
    modelo.add(Dense(1, activation='sigmoid'))

    modelo.compile(optimizer = 'adam', loss ='binary_crossentropy', metrics = ['accuracy'])

    modelo.fit(x_train, y_train, epochs = 150, batch_size = 10, validation_data = (x_test, y_test))

    return modelo
```

```python
modelo2 = KerasClassifier(build_fn = CriarRede, epochs = 100, batch_size = 10)
modelo2
```

```python
resultado = cross_val_score(estimator = modelo2, X = previsores, y = classe, cv = 10, scoring = 'accuracy')
resultado
```
