```python
from keras.models import Model, Sequential
import numpy as np
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Dense, Input
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

```python
x_train = x_train.reshape((-1, 784))
x_test = x_test.reshape((-1, 784))
```

```python
model = Sequential()

model.add(Dense(128, activation = 'relu', input_dim = x_train.shape[1]))

model.add(Dense(x_train.shape[1], activation = 'sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

model.fit(x_train, x_train, epochs=50, batch_size=256, validation_data=(x_test, x_test))
```

```python
dimensao = Input(shape = (x_train.shape[1],))
camada_encododer = model.layers[0]
encoder = Model(dimensao, camada_encododer(dimensao))
encoder.summary()
```

```python
x_train_cod = encoder.predict(x_train)
```

```python
x_test_cod = encoder.predict(x_test)
```

```python
classe_dummy_train = to_categorical(y_train)
classe_dummy_test = to_categorical(y_test)
```

```python
# Sem redução de dimensionalidade

model1 = Sequential()

model1.add(Dense(397, activation = 'relu', input_dim = x_train.shape[1]))

model1.add(Dense(397, activation = 'relu'))

model1.add(Dense(10, activation = 'softmax'))
```

```python
model1.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model1.fit(x_train, classe_dummy_train, batch_size = 256, epochs = 100, validation_data = (x_test, classe_dummy_test)) 
```

```python
# Com redução de dimensionalidade

model2 = Sequential()

model2.add(Dense(21, activation = 'relu', input_dim = 128))

model2.add(Dense(21, activation = 'relu'))

model2.add(Dense(10, activation = 'softmax'))
```

```python
model2.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
model2.fit(x_train_cod, classe_dummy_train, batch_size = 256, epochs = 200, validation_data = (x_test_cod, classe_dummy_test)) 
```
