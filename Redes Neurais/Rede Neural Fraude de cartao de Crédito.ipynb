```python
data = data.drop_duplicates()
```

```python
data.describe()
```

```python
data['Class'] = data['Class'].str.replace("'", '').astype(float)
```

```python
previsores = data.iloc[: , :-1].values
classe =data.iloc[: , -1].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
previsores = scaler.fit_transform(previsores)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(previsores, classe,
                                                  test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

forest = ExtraTreesClassifier()
forest.fit(x_train, y_train)
importancias = forest.feature_importances_
importancias
```

```python
x_train = x_train[:, [3, 4 , 9, 10, 11, 12, 16]]
x_test = x_test[: , [3, 4 , 9, 10, 11, 12, 16]]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()

model.add(Dense(16, kernel_initializer='random_uniform', input_dim=x_train.shape[1], activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs = 100,  batch_size=64, validation_data = (x_test, y_test))
```

```python
y_pred = model.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(y_test, y_pred)
matriz
```

```python
acertos = sum(matriz[i, i] for i in range(len(matriz)))
porcentagem_acertos = acertos / len(y_test)


print(f'Total de acertos na matriz de confusÃ£o: {acertos}')
print(f'Porcentagem de acertos: {porcentagem_acertos * 100:.2f}%')
```
