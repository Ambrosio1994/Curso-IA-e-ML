```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
import numpy as np
```

```python
df = datasets.load_iris()
previsores = df.data
classe = df.target
classe
```

```python
classe = to_categorical(classe)
classe
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores, classe, test_size=0.3, random_state=0)
```

```python
modelo = Sequential()

modelo.add(Dense(units = 5, input_dim = 4))

modelo.add(Dense(units = 4))

modelo.add(Dense(units = 3, activation = 'softmax'))
```

```python
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

modelo.fit(x_train, y_train, epochs = 1000, validation_data = (x_test, y_test))
```

```python
pred = modelo.predict(x_test)
pred = (pred > 0.5)
pred
```

```python
y_test_matrix = [np.argmax(t) for t in y_test]
y_pred_matrix = [np.argmax(t) for t in pred]
```

```python
matriz = confusion_matrix(y_test_matrix, y_pred_matrix)
matriz
```
