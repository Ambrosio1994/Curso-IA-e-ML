```python
import pandas as pd
```

```python
df_train = pd.read_csv('Downloads/petr4_treinamento_ex.csv')
df_train
```

```python
df_train.isna().sum()
```

```python
df_train.dtypes
```

```python
for coluna in df_train.columns:
    if df_train[coluna].dtype == 'float64':
        if df_train[coluna].isna().any():
            media = df_train[coluna].mean()
            df_train[coluna].fillna(media, inplace=True)

for coluna in df_train.columns:
    if df_train[coluna].dtype == 'object':
        if df_train[coluna].isna().any():
            moda = df_train[coluna].mode().iloc[0]
            df_train[coluna].fillna(moda, inplace=True)
```

```python
df_train.isna().sum()
```

```python
treino = df_train.iloc[:, 1:2].values
treino
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

treino = scaler.fit_transform(treino)
treino
```

```python
x_train = []  
y_train = []  
timestamp = 120
```

```python
for i in range(timestamp, len(treino)):
    x_train.append(treino[i - timestamp:i, 0])
    y_train.append(treino[i, 0])
```

```python
import numpy as np

x_train = np.array(x_train)
y_train = np.array(y_train)

x_train
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train
```

```python
from keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM, Dropout

model = Sequential()

model.add(LSTM(120, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(100, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(80, return_sequences=True))
model.add(Dropout(0.2))

model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(units=1))

model.compile(optimizer = 'rmsprop', loss ='mean_squared_error', metrics = ['mean_squared_error'])

model.summary()
```

```python
model.fit(x_train, y_train, epochs = 100, batch_size = 32)
```

```python
df_test = pd.read_csv('Downloads/petr4_teste_ex.csv')
df_test.head()
```

```python
y_test = df_test.iloc[:, 1:2].values
y_test
```

```python
base_completa = pd.concat((df_train['Open'], df_test['Open']), axis = 0)
base_completa
```

```python
entradas = base_completa[len(base_completa) - len(df_test) - timestamp:].values
entradas
```

```python
entradas = entradas.reshape(-1,1)
entradas
```

```python
entradas = scaler.transform(entradas)
entradas
```

```python
x_test = []

length = len(entradas)

for i in range(timestamp, length):
    
    x_test.append(entradas[i - timestamp:i, 0])

x_test = np.array(x_test)

x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)
```

```python
length
```

```python
pred = model.predict(x_test)
pred = scaler.inverse_transform(pred)
pred
```

```python
(pred - y_test).mean()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 15, 8

plt.plot(y_test, color='b', label='PreÃ§o real das aÃ§Ãµes')
plt.plot(pred, color='r', label='PrevisÃ£o das aÃ§Ãµes')

plt.title('PrevisÃ£o do PreÃ§o das AÃ§Ãµes')
plt.xlabel('Tempo')
plt.ylabel('PreÃ§o das AÃ§Ãµes')

plt.legend()


plt.show()
```
