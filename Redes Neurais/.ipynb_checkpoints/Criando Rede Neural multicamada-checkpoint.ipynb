```python
from sklearn import datasets
import numpy as np
```

```python
df = datasets.load_breast_cancer()
```

```python
entradas = df.data
valoressaida = df.target
saidas =  np.empty([569,1], dtype = int)
```

```python
for i in range(569):
    saidas[i] = valoressaida[i]
saidas
```

```python
pesos0 = 2 *  np.random.random((30,3)) - 1
pesos1 = 2 *  np.random.random((3,1)) - 1
```

```python
def sigmoid(soma):
    return 1 / (1+ np.exp(-soma))
```

```python
def sigmoidderivada(sig):
    return sig * (1 - sig)
```

```python
epoch = 1000
for i in range(epoch):
    camadaEntrada = entradas
    somasinapse0 = np.dot(camadaEntrada, pesos0)
    camadaoculta = sigmoid(somasinapse0).round(3)
    somasinapse1 = np.dot(camadaoculta, pesos1)
    camadasaida = sigmoid(somasinapse1).round(3)
    errosaida = saidas - camadasaida
    mediaerro = np.mean(abs(errosaida)).round(3)
    print(f'Erro: {mediaerro}')
    derivadasaida = sigmoidderivada(camadasaida).round(3)
    deltasaida = (errosaida * derivadasaida).round(3)
    pesos1t = pesos1.T
    deltacamadaoculta = (deltasaida.dot(pesos1t) * sigmoidderivada(camadaoculta)).round(3)
    camadaocultaT = camadaoculta.T
    pesosnovos1 = camadaocultaT.dot(deltasaida).round(3)
    taxadeaprendizagem = 0.7
    momento = 1
    pesos1 = ((pesos1 * momento) + (pesosnovos1 * taxadeaprendizagem)).round(3)
    camadaentradat = camadaEntrada.T
    pesosnovos0 = camadaentradat.dot(deltacamadaoculta)
    pesos0 = ((pesos0 * momento) + (pesosnovos0 * taxadeaprendizagem)).round(3)
```
