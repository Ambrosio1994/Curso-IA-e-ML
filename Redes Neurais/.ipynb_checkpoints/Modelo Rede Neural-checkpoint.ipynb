```python
from sklearn.ensemble import ExtraTreesClassifier
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
previsores = df.iloc[:,:-1].values
classe = df.iloc[:,-1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(previsores, classe, test_size = 0.3, random_state = 42)
```

```python
modelo = ExtraTreesClassifier()
modelo = modelo.fit(x_train,y_train)
```

```python
importancias = modelo.feature_importances_
importancias
```

```python
x_train = x_train[:,[0, 2, 6, 7, 20, 22, 23, 27]]
x_test = x_test[:,[0, 2, 6, 7, 20, 22, 23, 27]]
```

```python
modelo = Sequential()

modelo.add(Dense(16, kernel_initializer='random_uniform', input_dim = x_train.shape[1], activation='relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
modelo.add(Dropout(0.1))

modelo.add(Dense(1, activation='sigmoid'))
```

```python
modelo.compile(optimizer = 'adam', loss ='binary_crossentropy', metrics = ['accuracy'])

modelo.fit(x_train, y_train, epochs = 1000, batch_size = 20, validation_data = (x_test, y_test))
```

```python
pred = modelo.predict(x_test)
pred = (pred > 0.5)
pred
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)
cm
```

```python
def CriarRede():
    modelo = Sequential()

    modelo.add(Dense(16, kernel_initializer='random_uniform', input_dim = x_train.shape[1], activation='relu'))
    modelo.add(Dropout(0.3))
    
    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    modelo.add(Dropout(0.2))
    
    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    modelo.add(Dropout(0.1))

    modelo.add(Dense(16, kernel_initializer='random_uniform', activation='relu'))
    
    modelo.add(Dense(1, activation='sigmoid'))

    modelo.compile(optimizer = 'adam', loss ='binary_crossentropy', metrics = ['accuracy'])

    modelo.fit(x_train, y_train, epochs = 100, batch_size = 20, validation_data = (x_test, y_test))

    return modelo
```

```python
from sklearn.model_selection import cross_val_score
from scikeras.wrappers import KerasClassifier
```

```python
modelo2 = KerasClassifier(build_fn = CriarRede, epochs = 100, batch_size = 10)
modelo2
```

```python
resultado = cross_val_score(estimator = modelo2, X = previsores, y = classe, cv = 10, scoring = 'accuracy')
resultado
```
