```python
import pandas as pd
df = pd.read_csv('personagens.csv')
df
```

```python
previsores = df.iloc[:, :-1].values
classe = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(previsores, classe, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
import tensorflow as tf
from keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
modelo = Sequential()

modelo.add(Dense(16, activation = 'relu', input_dim = x_train.shape[1]))
modelo.add(Dropout(0.2))

modelo.add(Dense(16, activation = 'relu'))
modelo.add(Dropout(0.2))

modelo.add(Dense(1, activation='sigmoid'))
```

```python
modelo.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
modelo.fit(x_train, y_train, batch_size = 10, epochs = 100, validation_data = (x_test, y_test))
```

```python
pred = modelo.predict(x_test)
pred 
```
