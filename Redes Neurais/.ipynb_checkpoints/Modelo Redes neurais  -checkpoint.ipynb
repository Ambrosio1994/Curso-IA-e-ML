```python
import pandas as pd
from keras.models import Model
from keras.layers import Dense, Dropout, Activation, Input
```

```python
df = pd.read_csv('games.csv')
df
```

```python
df.columns
```

```python
df = df.drop([ 'Other_Sales', 'Global_Sales',  'Developer'], axis = 1)
```

```python
df.isna().sum()
```

```python
for coluna in df.columns:
    if df[coluna].dtype == 'float64':
        if df[coluna].isna().any():
            media = df[coluna].mean()
            df[coluna].fillna(media, inplace=True)

for coluna in df.columns:
    if df[coluna].dtype == 'object':
        if df[coluna].isna().any():
            moda = df[coluna].mode().iloc[0]
            df[coluna].fillna(moda, inplace=True)
```

```python
df.loc[df['NA_Sales'] < 1] = df['NA_Sales'].mean()
df.loc[df['EU_Sales'] < 1] = df['EU_Sales'].mean()
```

```python
df = df.drop('Name', axis = 1)
df
```

```python
previsores = df.iloc[:, [0, 1, 2, 3, 7, 8, 9, 10, 11]].values
vendas_na = df.iloc[: , 4].values
vendas_eu = df.iloc[: , 5].values
vendas_jp = df.iloc[: , 6].values
```

```python
labelencoder = LabelEncoder()

for coluna in range(previsores.shape[1]):
    if previsores[:, coluna].dtype == object:
        
        previsores[:, coluna] = previsores[:, coluna].astype(str)
        
        previsores[:, coluna] = labelencoder.fit_transform(previsores[:, coluna])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Create a ColumnTransformer object.
column_transformer = ColumnTransformer([
    ('categorical_features', OneHotEncoder(), [0, 2, 3, 8]),
])

# Transform the DataFrame.
previsores = column_transformer.fit_transform(previsores)
```

```python
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler()
previsores = scaler.fit_transform(previsores)
```

```python
entrada = Input(shape = (previsores.shape[1], ))
camada_oculta1 = Dense(units = 60, activation = 'sigmoid')(entrada)
camada_oculta2 = Dense(units = 60, activation = 'sigmoid')(camada_oculta1)
saida1 = Dense(units = 1, activation = 'linear')(camada_oculta2)
saida2 = Dense(units = 1, activation = 'linear')(camada_oculta2)
saida3 = Dense(units = 1, activation = 'linear')(camada_oculta2)
```

```python
modelo = Model(inputs = entrada, outputs = [saida1, saida2, saida3])
modelo.compile(optimizer = 'adam', loss = 'mse')
modelo.fit(previsores,[vendas_na, vendas_eu, vendas_jp], batch_size = 5000, epochs = 100)
```
