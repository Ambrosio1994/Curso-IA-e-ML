```python
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
previsores = df.iloc[:, :-1].values
classe = df.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

classe = labelencoder.fit_transform(classe)
```

```python
from keras.utils import to_categorical

classe_dummy = to_categorical(classe)
classe_dummy
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(previsores, classe_dummy, test_size = 0.3, random_state = 42)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
modelo = Sequential()

modelo.add(Dense(5, activation = 'relu', input_dim= x_train.shape[1]))
modelo.add(Dropout(0.1))

modelo.add(Dense(4, activation = 'relu'))

modelo.add(Dense(3, activation = 'softmax'))
```

```python
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
modelo.fit(x_train, y_train, batch_size=20, epochs=1000)
```

```python
pred = modelo.predict(x_test)
pred = (pred > 0.5)
```

```python
import numpy as np

y_test_matrix = [np.argmax(t) for t in y_test]
y_pred_matrix = [np.argmax(t) for t in pred]
```

```python
from sklearn.metrics import confusion_matrix

matriz = confusion_matrix(y_test_matrix, y_pred_matrix)
matriz
```

```python
def criar_modelo():
    modelo = Sequential()

    modelo.add(Dense(5, activation = 'relu', input_dim= x_train.shape[1]))

    modelo.add(Dense(4, activation = 'relu'))
    
    modelo.add(Dense(3, activation = 'softmax'))

    modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])

    return modelo
```

```python
modelos = []
```

```python
from scikeras.wrappers import KerasClassifier

modelo = KerasClassifier(model = criar_modelo, epochs = 100, batch_size = 10)
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(estimator = modelo, X = previsores, y = classe_dummy, cv = 10, scoring = 'accuracy')
```
