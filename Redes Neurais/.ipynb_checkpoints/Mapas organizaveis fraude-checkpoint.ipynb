```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from minisom import MiniSom
from pylab import pcolor, colorbar, plot
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Downloads/credit_data.csv')
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df.loc[df['age'] < 0, 'age'] = df['age'].mean()
df.head()
```

```python
scaler = MinMaxScaler()
```

```python
previsores = df.iloc[:, 0:-1].values
previsores = scaler.fit_transform(previsores)
previsores
```

```python
classe = df.iloc[:, -1].values
classe
```

```python
som = MiniSom(x = 15, y = 15, input_len = previsores.shape[1], random_seed = 0)
som.random_weights_init(previsores)
som.train_random(data = previsores,num_iteration = 100)
```

```python
pcolor(som.distance_map().T)
colorbar()
```

```python
markers = ['o', 's']
color = ['r', 'g']
```

```python
plt.figure(figsize=(10, 10))
ax = plt.gca()
ax.set_facecolor('gray')


for i, x in enumerate(previsores):
    w = som.winner(x)
    plot(w[0] + 0.5, w[1] + 0.5, markers[classe[i]], markerfacecolor = 'None',
         markersize = 10, markeredgecolor = color[classe[i]], markeredgewidth = 2)
```
