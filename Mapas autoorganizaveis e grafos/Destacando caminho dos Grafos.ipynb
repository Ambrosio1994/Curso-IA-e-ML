```python
from igraph import Graph
from igraph import plot
```

```python
grafo = Graph(edges = [(0,2), (0,1), (1,4), (1,5), (2,3), (6,7), (3,7), (4,7),(5,6)], directed = True)
grafo.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
grafo.es['weight'] = [2,1,2,1,2,1,3,1]
plot(grafo, bbox = (0,0,300,300), edge_label = grafo.es['weight'])
```

```python
caminho_vertice = grafo.get_shortest_paths(0,7, output = 'vpath')
for i in caminho_vertice[0]:
    print(grafo.vs[i]['label'])
```

```python
caminho_arresta = grafo.get_shortest_paths(0,7, output = 'epath')
caminho_arresta
```

```python
caminho_arresta_id = []
for i in caminho_arresta[0]:
    caminho_arresta_id.append(i)
caminho_arresta_id

distancia = 0
for e in grafo.es:
    if e.index in caminho_arresta_id:
        distancia += grafo.es[e.index]['weight']
distancia
```

```python
for v in grafo.vs:
    if v['label'] in caminho_vertice:
        v['color'] = 'green'
    else:
        v['color'] = 'gray'
```

```python
for e in grafo.es:
    if e.index in caminho_arresta_id:
        e['color'] = 'green'
    else:
        e['color'] = 'gray'
```

```python
plot(grafo, bbox = (0, 0, 300, 300))
```
