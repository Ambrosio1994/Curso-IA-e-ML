Importação dos pacotes

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as ex

from sklearn.neighbors import NearestNeighbors

import warnings
warnings.filterwarnings('ignore')
```

Importação dos Dataframes

```python
data_books = pd.read_csv('Books.csv')
data_ratings = pd.read_csv('Ratings.csv')
data_user = pd.read_csv('Users.csv')
```

```python
data_books.head(2)
```

```python
data_ratings.head(2)
```

```python
data = data_books.merge(data_ratings, how='inner', on='ISBN')
data.head(2)
```

```python
data = data.merge(data_user, how='inner', on='User-ID')
data.head(2)
```

```python
data['Year-Of-Publication'] = pd.to_numeric(data['Year-Of-Publication'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data.info()
```

```python
data[data['Year-Of-Publication'].apply(lambda x: isinstance(x, str))] = ''
data.info()
```

```python
data.head()
```

```python
data['Location'] = data['Location'].str.split(',').str[-1]
data.head()
```

```python
data = data.loc[data['Book-Rating'] > 0]
data.describe()
```

```python
data = data.loc[data['Year-Of-Publication'] >= 1965]
data = data.loc[data['Year-Of-Publication'] <= 2023]
data = data.loc[data['Age'] <= 90]
data = data.loc[data['Age'] >= 7]
```

```python
data['Year-Of-Publication'] = data['Year-Of-Publication'].astype(int)
data.info()
```

```python
sampled_data = data.sample(n=10000)

matriz = sampled_data.pivot_table(values='Book-Rating', index='Book-Title', columns='User-ID')
matriz.fillna(0, inplace=True)

from scipy.sparse import csc_matrix
matriz_esparsa = csc_matrix(matriz)
```

```python
matriz
```

```python
model = NearestNeighbors(n_neighbors=5,
                         algorithm='brute',
                         metric="minkowski")
model.fit(matriz_esparsa)
```

```python
for posicao, titulo in enumerate(matriz.index):
    if 'Harry Potter' in str(titulo):
        print(posicao, titulo)
```

```python
livro = matriz_esparsa[2713, :].reshape(1, -1)
```

```python
distancia, recomendacao = model.kneighbors(livro)
```
